```python
import random
import math
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import matplotlib.patches as mpatches

p = []
p.append(float(0.1))
p.append(float(0.3))
p.append(float(0.4))
n = len(p)

n_show = 5000  #  
n_start = 50  #  


# ------------------ -----------------------

#   
def show_page(number_page, reward_show, number_show, p_show):
    p = p_show
    reward_show[number_page] += bernoulli.rvs(p[number_page])
    number_show[number_page] += 1
    return 0

#   
#    
def regret(par_list):
	#    
	temp_best_index_page = p.index(max(p))
	temp_regret_list = [0 for i in range(n_show)]
	for i in range(n_show):
		j, temp_var = 0, 0
		while j < i:
			temp_var += p[par_list[j]]
			j += 1
		temp_regret_list[i] = i * p[temp_best_index_page] - temp_var
	return temp_regret_list


# ---------------------Thompson Sampling--------------------

sum_regret_list_sampling = [0 for i in range(n_show)]
sum_reward_show_sampling = [0 for i in range(n)]  #    
sum_number_show_sampling = [1 for i in range(n)]  #   
sum_ctr_sampling = [0 for i in range(n)]  # click-through-rate
sum_mean_ctr_list_sampling = [0 for i in range(n_show)]


def thompson_sampling(u):
	#   ctr
	for i in range(n):
		ctr_sampling[i] = reward_show_sampling[i] / number_show_sampling[i]

	#     max
	maximum, best_page_index, i = 0, 0, 0
	while i < n:
		beta_rasp = random.betavariate(reward_show_sampling[i] + 1, number_show_sampling[i] - reward_show_sampling[i] + 1)
		if beta_rasp > maximum and beta_rasp < u:
			maximum = beta_rasp
			best_page_index = i
		i += 1
	return best_page_index



# ---------------------Thompson Sampling 2--------------------

sum_regret_list_sampling_2 = [0 for i in range(n_show)]
sum_reward_show_sampling_2 = [0 for i in range(n)]  #    
sum_number_show_sampling_2 = [1 for i in range(n)]  #   
sum_ctr_sampling_2 = [0 for i in range(n)]  # click-through-rate
sum_mean_ctr_list_sampling_2 = [0 for i in range(n_show)]


def thompson_sampling_2(u_2):
	#   ctr
	for i in range(n):
		ctr_sampling_2[i] = reward_show_sampling_2[i] / number_show_sampling_2[i]

	#     max
	maximum, best_page_index_2, i = 0, 0, 0
	while i < n:
		beta_rasp_2 = random.betavariate(reward_show_sampling_2[i] + 1, number_show_sampling_2[i] - reward_show_sampling_2[i] + 1)
		if beta_rasp_2 > maximum and beta_rasp_2 < u_2:
			maximum = beta_rasp_2
			best_page_index_2 = i
		i += 1
	return best_page_index_2


# ---------------------Thompson Sampling 3--------------------

sum_regret_list_sampling_3 = [0 for i in range(n_show)]
sum_reward_show_sampling_3 = [0 for i in range(n)]  #    
sum_number_show_sampling_3 = [1 for i in range(n)]  #   
sum_ctr_sampling_3 = [0 for i in range(n)]  # click-through-rate
sum_mean_ctr_list_sampling_3 = [0 for i in range(n_show)]


def thompson_sampling_3(u_3):
	#   ctr
	for i in range(n):
		ctr_sampling_3[i] = reward_show_sampling_3[i] / number_show_sampling_3[i]

	#     max
	maximum, best_page_index_3, i = 0, 0, 0
	while i < n:
		beta_rasp_3 = random.betavariate(reward_show_sampling_3[i] + 1, number_show_sampling_3[i] - reward_show_sampling_3[i] + 1)
		if beta_rasp_3 > maximum and beta_rasp_3 < u_3:
			maximum = beta_rasp_3
			best_page_index_3 = i
		i += 1
	return best_page_index_3



#  
#     
n_count = 0
while n_count < n_start:

    #  Thompson Sampling
    reward_show_sampling = [0 for i in range(n)]  #    
    number_show_sampling = [1 for i in range(n)]  #   
    ctr_sampling = [0 for i in range(n)]  # click-through-rate
    page_in_time_sampling = [0 for i in range(n_show)]
    mean_ctr_list_sampling = [0 for i in range(n_show)]
    
    #  Thompson Sampling 2
    reward_show_sampling_2 = [0 for i in range(n)]  #    
    number_show_sampling_2 = [1 for i in range(n)]  #   
    ctr_sampling_2 = [0 for i in range(n)]  # click-through-rate
    page_in_time_sampling_2 = [0 for i in range(n_show)]
    mean_ctr_list_sampling_2 = [0 for i in range(n_show)]
    
    #  Thompson Sampling 3
    reward_show_sampling_3 = [0 for i in range(n)]  #    
    number_show_sampling_3 = [1 for i in range(n)]  #   
    ctr_sampling_3 = [0 for i in range(n)]  # click-through-rate
    page_in_time_sampling_3 = [0 for i in range(n_show)]
    mean_ctr_list_sampling_3 = [0 for i in range(n_show)]

    t = 1
    while t < n_show:
        if t == 1:
            page_index = 0
            page_index_2 = 0
            page_index_3 = 0
        if t == 2:
            page_index = 1
            page_index_2 = 1
            page_index_3 = 1
        if t == 3:
            page_index = 2
            page_index_2 = 2
            page_index_3 = 2

        p_show = p


        # Thompson Sampling
        mean_ctr = 0
        u = 0.5
        if t > 3:
            page_index = thompson_sampling(u)
        page_in_time_sampling[t] = page_index
        show_page(page_index, reward_show_sampling, number_show_sampling, p_show)

        for i in range(n):
            mean_ctr += reward_show_sampling[i]

        mean_ctr_list_sampling[t] = mean_ctr / t


        # Thompson Sampling 2
        mean_ctr_2 = 0
        u_2 = 0.7
        if t > 3:
            page_index_2 = thompson_sampling_2(u_2)
        page_in_time_sampling_2[t] = page_index_2
        show_page(page_index_2, reward_show_sampling_2, number_show_sampling_2, p_show)

        for i in range(n):
            mean_ctr_2 += reward_show_sampling_2[i]

        mean_ctr_list_sampling_2[t] = mean_ctr_2 / t
        
        # Thompson Sampling 3
        mean_ctr_3 = 0
        u_3 = 1
        if t > 3:
            page_index_3 = thompson_sampling_3(u_3)
        page_in_time_sampling_3[t] = page_index_3
        show_page(page_index_3, reward_show_sampling_3, number_show_sampling_3, p_show)

        for i in range(n):
            mean_ctr_3 += reward_show_sampling_3[i]

        mean_ctr_list_sampling_3[t] = mean_ctr_3 / t


        t += 1

    #   
    regret_list_sampling = regret(page_in_time_sampling)
    regret_list_sampling_2 = regret(page_in_time_sampling_2)
    regret_list_sampling_3 = regret(page_in_time_sampling_3)

	#    
    for i in range(n_show):
        sum_regret_list_sampling[i] += regret_list_sampling[i]
        sum_regret_list_sampling_2[i] += regret_list_sampling_2[i]
        sum_regret_list_sampling_3[i] += regret_list_sampling_3[i]

        #   ctr
        sum_mean_ctr_list_sampling[i] += mean_ctr_list_sampling[i]
        sum_mean_ctr_list_sampling_2[i] += mean_ctr_list_sampling_2[i]
        sum_mean_ctr_list_sampling_3[i] += mean_ctr_list_sampling_3[i]

    for i in range(n):
        # thompson sampling
        sum_number_show_sampling[i] += number_show_sampling[i]
        sum_reward_show_sampling[i] += reward_show_sampling[i]
        sum_ctr_sampling[i] += ctr_sampling[i]
        
        # thompson sampling
        sum_number_show_sampling_2[i] += number_show_sampling_2[i]
        sum_reward_show_sampling_2[i] += reward_show_sampling_2[i]
        sum_ctr_sampling_2[i] += ctr_sampling_2[i]
        
        # thompson sampling
        sum_number_show_sampling_3[i] += number_show_sampling_3[i]
        sum_reward_show_sampling_3[i] += reward_show_sampling_3[i]
        sum_ctr_sampling_3[i] += ctr_sampling_3[i]

    n_count += 1


for i in range(n_show):
	# regret
    sum_regret_list_sampling[i] /= n_start
    sum_regret_list_sampling_2[i] /= n_start
    sum_regret_list_sampling_3[i] /= n_start

    # ctr
    sum_mean_ctr_list_sampling[i] /= n_start
    sum_mean_ctr_list_sampling_2[i] /= n_start
    sum_mean_ctr_list_sampling_3[i] /= n_start


for i in range(n):
    sum_reward_show_sampling[i] /= n_start
    sum_number_show_sampling[i] /= n_start
    sum_ctr_sampling[i] /= n_start

    sum_reward_show_sampling_2[i] /= n_start
    sum_number_show_sampling_2[i] /= n_start
    sum_ctr_sampling_2[i] /= n_start

    sum_reward_show_sampling_3[i] /= n_start
    sum_number_show_sampling_3[i] /= n_start
    sum_ctr_sampling_3[i] /= n_start


#   


time = [i for i in range(n_show)]
plt.figure(1)
plt.plot(time, sum_regret_list_sampling, linestyle='--', color='orange', label='  (u = 0.5)')
plt.plot(time, sum_regret_list_sampling_2, linestyle='--', color='blue', label='  (u = 0.7)')
plt.plot(time, sum_regret_list_sampling_3, linestyle='--', color='green', label='  (u = 1)')
plt.title('  ', fontsize=18)
plt.xlabel('', fontsize=16)
plt.ylabel('', fontsize=16)
plt.legend(loc='upper left')
plt.grid(True)
#plt.savefig('__6.png')
plt.show()

#  
plt.figure(2)
plt.plot(time, sum_mean_ctr_list_sampling, linestyle='--', color='orange', label='  (u = 0.5)')
plt.plot(time, sum_mean_ctr_list_sampling_2, linestyle='--', color='blue', label='  (u = 0.7)')
plt.plot(time, sum_mean_ctr_list_sampling_3, linestyle='--', color='green', label='  (u = 1)')

plt.title('', fontsize=18)
plt.xlabel('', fontsize=16)
plt.ylabel('', fontsize=16)
plt.legend(loc='lower right')
plt.grid(True)
#plt.savefig('__6.png')
plt.show()
```
